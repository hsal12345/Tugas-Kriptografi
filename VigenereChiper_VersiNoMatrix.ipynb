```python
# Vigenere Criptor
import numpy as np

letter = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'

def vigenere_kriptor(plain, key):
    
    plain = plain.upper()
    key = key.upper()
    
    #Buat Plain key dari key yang diberikan
    plain_key = ''
    if len(plain) <= len(key):
        for k in range(len(plain)):
            if plain[k] != ' ':
                plain_key += key[k]
            else:
                plain_key += ' '
    else:
        looper = 0
        for k2 in range(len(plain)):
            if plain[k2] != ' ':
                if looper > len(key) - 1:
                    plain_key += key[0]
                    looper = 1
                    print('k : ',k2)
                    #print('key : ',len(key))
                    print('looper : ',looper)
                else:
                    plain_key += key[looper]
                    looper += 1
                    print('k : ',k2)
                    #print('key : ',len(key))
                    print('looper : ',looper)
            else:
                plain_key += ' '
                
    # vigenere kriptor
    chiper_text = ''
    for x in range(len(plain)):
        if plain[x] != ' ':
            c = letter.find(plain[x]) + letter.find(plain_key[x])
            c = np.mod(c, 26)
            
            chiper_text += letter[c]
        else:
            chiper_text += ' '
            
    print('plain text   : ',plain)
    print('key to plain : ',plain_key)
        
    return chiper_text
      
        
vigenere_kriptor('plain text adalah text yang menggunkan', 'inilah')
```

```python
# Dekriptor Vigenere

def vigenere_dekriptor(chiper, key):
    
    chiper = chiper.upper()
    key = key.upper()
    
    #Buat Plain key dari key yang diberikan
    plain_key = ''
    if len(chiper) <= len(key):
        for k in range(len(chiper)):
            if chiper[k] != ' ':
                plain_key += key[k]
            else:
                plain_key += ' '
    else:
        looper = 0
        for k2 in range(len(chiper)):
            if chiper[k2] != ' ':
                if looper > len(key) - 1:
                    plain_key += key[0]
                    looper = 1
                    print('k : ',k2)
                    #print('key : ',len(key))
                    print('looper : ',looper)
                else:
                    plain_key += key[looper]
                    looper += 1
                    print('k : ',k2)
                    #print('key : ',len(key))
                    print('looper : ',looper)
            else:
                plain_key += ' '
                
    # vigenere kriptor
    plain_text = ''
    for x in range(len(chiper)):
        if chiper[x] != ' ':
            c = letter.find(chiper[x]) - letter.find(plain_key[x])
            c = np.mod(c, 26)
            
            plain_text += letter[c]
        else:
            plain_text += ' '
            
    print('chiper text   : ',chiper)
    print('key to plain : ',plain_key)
        
    return plain_text

vigenere_dekriptor('XYITN AMKB LDHTNP EEEB LIYG TMAORUUSNV', 'inilah')
```
