```python
#Kelompok B
#Nomor 1b

import numpy as np

letter = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
def caesarChiper(k, plain):
    chiperText = ''
    plain = plain.upper()
    for x in plain:
        if x != " ":
            c = letter.find(x) + k
            c = np.mod(c, 26)
            
            chiperText = chiperText + letter[c]
        else:
            chiperText = chiperText + plain[plain.find(x)]
        
    return chiperText
        
print('1b :', caesarChiper(13, 'DO NOT PASS GO'))

#Kelompok B
#Nomor 4b dan 9

def caesarBreaker1(chiperT):
    chiperT = chiperT.upper()
    for i in range(26):
        plainText = ''
        for x in chiperT:
            if x != " ":
                p = letter.find(x) - i
                p = np.mod(p, 26)

                plainText = plainText + letter[p]
            else:
                plainText = plainText + chiperT[chiperT.find(x)]

        print(i,' : ',plainText)

print('\n4b : ')
caesarBreaker1('WHVW WRGDB')
print('\n9 : ')
caesarBreaker1('ERC WYJJMGMIRXPC EHZERGIH XIGLRSPSKC MW MRHMWXMRKYMWLEFPI JVSQ QEKMG')
```
